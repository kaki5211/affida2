```python
import pykakasi

def convert_to_romaji(text):
    kakasi = pykakasi.kakasi()
    kakasi.setMode("H", "a")        # Hiragana to ascii
    kakasi.setMode("K", "a")        # Katakana to ascii
    kakasi.setMode("J", "a")        # Japanese to ascii
    kakasi.setMode("r", "Hepburn")  # use Hepburn Roman table
    kakasi.setMode("s", True)       # add space
    kakasi.setMode("C", False)      # no capitalize
    conv = kakasi.getConverter()
    result = conv.do(text)
    return result

def get_json_data(soup, kind):
    # NAME_INFO 取得
    if kind == "actor":
        name_info = soup.xxxx
    elif kind in ["maker", "label", "series", "tag"]:
        name_info = soup.xxxx
    else:
        print("'kind' が見つかりません")
        return
    
    # NAME_ENG_INFO 取得
    name_eng_info = convert_to_romaji(name_info)
    
    # BIRTH_INFO 取得
    if kind == "actor":
        birth_info = soup.xxxx
    
    # JSON DATA作成
    if kind == "actor":
        json_str = json.dumps({
        "id": None,
        "name": name_info,
        "birth": birth_info,
        "name_eng": name_eng_info
        }, indent=4, ensure_ascii=False)
    elif kind in ["maker", "label", "series", "tag"]:
        json_str = json.dumps({
        "id": None,
        "name": name_info,
        "name_eng": name_eng_info
        }, indent=4, ensure_ascii=False)
    return json_str

result_str = get_json_data(soup, "actor")

    
```
